```python
#importing libraries

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
#loading the data
df = pd.read_csv("https://media.githubusercontent.com/media/praveenbakthavel/ML/refs/heads/main/Datasets/AmazonStockData.csv")

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.drop(columns=['Adj Close'])

df.head()
```

```python
df2 = df.drop(columns=['Date'])
df2.head()
```

```python
std_scaler = StandardScaler()

df_scaled = std_scaler.fit_transform(df2.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=[
  'Open','High','Low','Close','Volume'])

print("Scaled Dataset Using StandardScaler")
df_scaled.head()
```

```python
df_scaled['Date'] = df['Date']
df_scaled.head()
```

```python
df_scaled.info()
```

```python
x = df_scaled.drop(columns=['Close'])

x['Date'] = pd.to_datetime(x['Date'])

x["year"] = x["Date"].dt.year
x["month"] = x["Date"].dt.month
x["day"] = x["Date"].dt.day
x["dayofweek"] = x["Date"].dt.dayofweek

x = x.drop(columns=['Date'])

y = df_scaled['Close']

print(x.head())
print(y.head())
```

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
regr = LinearRegression()

regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
import pickle as pk
```

```python
FileName = "LinearRegressionModel.sav"
FileName2 = "df_scaled.sav"
```

```python
pk.dump(regr, open(FileName, 'wb') )
pk.dump(df_scaled, open(FileName2, 'wb') )
```

```python
load_model = pk.load(open(FileName, 'rb'))
```

```python
y_pred = load_model.predict(X_test)

print(y_pred)
```

```python
X_test['Close'] = y_pred

print(X_test.head())
```

```python
# Correct column selection
cols = ['Open', 'High', 'Low', 'Volume', 'Close']

# Inverse transform
y_pred_original = std_scaler.inverse_transform(X_test[cols].values)

# Convert back to DataFrame for readability
y_pred_original = pd.DataFrame(y_pred_original, columns=cols, index=X_test.index)

# Show results
y_pred_original.head()
```
